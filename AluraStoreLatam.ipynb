### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
Total_Ventas_Tienda1 = int(sum(tienda.Precio))
Total_Ventas_Tienda2 = int(sum(tienda2.Precio))
Total_Ventas_Tienda3 = int(sum(tienda3.Precio))
Total_Ventas_Tienda4 = int(sum(tienda4.Precio))
```

```python
print(f'El ingreso total de la tienda 1 es {int(Total_Ventas_Tienda1)}.')
print(f'El ingreso total de la tienda 2 es {int(Total_Ventas_Tienda2)}.')
print(f'El ingreso total de la tienda 3 es {int(Total_Ventas_Tienda3)}.')
print(f'El ingreso total de la tienda 4 es {int(Total_Ventas_Tienda4)}.')
```

# 2. Ventas por categoría

```python
def ventasporcategoria(tienda, nombretienda):

  df = pd.DataFrame(tienda)

  agrupar = df.groupby('Categoría del Producto')
  cantidades = [( x[0],len(x[1]['Categoría del Producto']) ) for x in agrupar]

  print(f'"{nombretienda.upper()}"')
  print([f'La cantidad de {y[0]} vendidos en la {nombretienda} son: {y[1]}' for y in cantidades])
  print(f'La categoría más popular de la {nombretienda} es: {max(cantidades)[0]}, y se vendió un total de {max(cantidades)[1]}')
  print('')

ventasporcategoria(tienda, 'tienda 1')
ventasporcategoria(tienda2, 'tienda 2')
ventasporcategoria(tienda3, 'tienda 3')
ventasporcategoria(tienda4, 'tienda 4')
```

# 3. Calificación promedio de la tienda


```python
def PromedioPorTienda(tienda, nombre):
  tienda['Calificación']
  promedio_calificacion = round(sum(tienda['Calificación'] / len(tienda['Calificación'])),2)
  print(f'La calificación promedio de la {nombre} es: {promedio_calificacion}')

PromedioPorTienda(tienda, 'tienda 1')
PromedioPorTienda(tienda2, 'tienda 2')
PromedioPorTienda(tienda3, 'tienda 3')
PromedioPorTienda(tienda4, 'tienda 4')
```

# 4. Productos más y menos vendidos

```python
def ventasporcategoria(tienda, nombretienda):

  df = pd.DataFrame(tienda)

  agrupar = df.groupby('Producto')
  cantidades = [( x[0], len(x[1]['Producto']) ) for x in agrupar]

  print(f'"{nombretienda.upper()}"')
  print([f'La cantidad de {y[0]} vendidos en la {nombretienda} son: {y[1]}' for y in cantidades])
  print(f'La categoría más popular de la {nombretienda} es: {max(cantidades)[0]}, y se vendió un total de {max(cantidades)[1]} unidades.')
  print(f'La categoría menos popular de la {nombretienda} es: {min(cantidades)[0]}, y se vendió un total de {min(cantidades)[1]} unidades.')
  print('')

ventasporcategoria(tienda, 'tienda 1')
ventasporcategoria(tienda2, 'tienda 2')
ventasporcategoria(tienda3, 'tienda 3')
ventasporcategoria(tienda4, 'tienda 4')
```

# 5. Envío promedio por tienda

```python
def promedioenvio (tienda, nombretienda):
  promedio_envio = round(( sum(tienda['Costo de envío']) / len(tienda['Costo de envío'])  ),2)
  print(f'El coste promedio de envío de la {nombretienda} es: {promedio_envio}.')

promedioenvio(tienda, 'tienda 1')
promedioenvio(tienda2, 'tienda 2')
promedioenvio(tienda3, 'tienda 3')
promedioenvio(tienda4, 'tienda 4')
```
